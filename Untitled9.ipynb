```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Read your .data file into a pandas DataFrame


# Assign column names
# Example of renaming columns properly
columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 
              'Feature6', 'Feature7', 'Feature8', 'Feature9', 'Feature10', 'class']
data = pd.read_csv(r"C:\Users\rodyd\Downloads\magic04.data",delimiter=',', header=None, names=columns)

# Now the data has meaningful column names
print(data.head())

```

```python

# Separate the dataset into two classes: gamma (signal) and hadron (background)
gammas = data[data['class'] == 'g']  # Gamma class
hadrons = data[data['class'] == 'h']  # Hadron class

# Balance the dataset by randomly sampling the gamma class to match hadron class size
gammas_balanced = gammas.sample(len(hadrons), random_state=42)

# Concatenate the balanced gamma events with the hadron events
df_balanced = pd.concat([gammas_balanced, hadrons])

# Shuffle the dataset to ensure randomness
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the distribution of classes in the balanced dataset
print(df_balanced['class'].value_counts())


```

```python

#by using sklearn.model_selection 
# Assuming you already have the balanced dataframe `df_balanced`
# First, split the data into a training set (70%) and the combined validation & test set (30%)
#The random_state is a seed for the random number generator used by train_test_split (42)
train_df, temp_df = train_test_split(df_balanced, test_size=0.3, random_state=42)#30 percent set aside(to use for val and test)

# Then, split the temp_df (combined validation & test set) into validation (15%) and test (15%)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Prepare the features (X) and target (y) for each split
X_train = train_df.drop('class', axis=1)
y_train = train_df['class']

X_val = val_df.drop('class', axis=1)
y_val = val_df['class']

X_test = test_df.drop('class', axis=1)
y_test = test_df['class']


#  train_df for training (70%)
#  val_df for validation (15%) helps in selecting the best k
#  test_df for testing (15%)
```

```python
print(data.columns)  # Check column names
```

```python
#accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

k_values = [1, 3, 5, 7, 2]
results = []

for k in k_values:
    # Train the K-NN model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Make predictions on the validation set
    y_pred = knn.predict(X_val)
    
    accuracy = accuracy_score(y_val, y_pred) #accuracy is  evaluation and predicted
#  (assuming 'g' is the positive class)
    precision = precision_score(y_val, y_pred, pos_label='g')
    recall = recall_score(y_val, y_pred, pos_label='g')
    f1 = f1_score(y_val, y_pred, pos_label='g')
# Confusion matrix
    conf_matrix = confusion_matrix(y_val, y_pred,labels=['g', 'h'])

   
    
    # Store results for comparison
    results.append({
        'k': k,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix': conf_matrix
    })
```

```python
results_df = pd.DataFrame(results)
print(results_df)

# Step 5: Visualize Confusion Matrices for each k
for result in results:
    cm = result['Confusion Matrix']
    sns.heatmap(cm, annot=True, fmt='d', cmap='pink', xticklabels=['Hadron', 'Gamma'], yticklabels=['Hadron', 'Gamma'])
    plt.title(f'Confusion Matrix for k={result["k"]}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Step 6: Evaluation on the Test Set with the Best Model
best_k = results_df.loc[results_df['F1-Score'].idxmax()]['k'] #the best k is the one with highest fscore
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

# Test the best model on the test set
y_test_pred = best_knn.predict(X_test)

# Evaluate on the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred,pos_label='g')
test_recall = recall_score(y_test, y_test_pred,pos_label='g')
test_f1 = f1_score(y_test, y_test_pred,pos_label='g')

print(f"Test Set Performance for k={best_k}:")
print(f"Accuracy: {test_accuracy:.4f}")
print(f"Precision: {test_precision:.4f}")
print(f"Recall: {test_recall:.4f}")
print(f"F1-Score: {test_f1:.4f}")
```
